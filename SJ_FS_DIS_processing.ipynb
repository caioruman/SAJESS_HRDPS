````python
yr = 2020
mnth = ["December","12"]
days = pd.Period(f'{yr}-{mnth[1]}-01').days_in_month
ind = pd.date_range(f'{int(yr)}-{int(mnth[1])}-01 00:00:00',f'{int(yr)}-{int(mnth[1])}-{days} 23:59:00',freq='T')
````

````python
path = f'../FixedStation/parsivel/{yr}_{mnth[1]}/raw/'
````

````python
files = [filename for filename in os.listdir(path) if filename.endswith('.txt')]
files.sort()
````

````python
listRowsSkip = []
for j in range(100000):
    if j not in range(1,100000,3):
        listRowsSkip.append(j)
````

````python
df = [pd.read_csv(path+file, header = None,sep = ';',skiprows = listRowsSkip, names = header, engine='python') for file in files]
````

````python
df_Parsivel = pd.concat(df)
df_Parsivel.drop(['toDelete'], axis = 1, inplace = True)
df_Parsivel['V0D0'] = np.nan
````

````python
df_Parsivel.reset_index(drop = True, inplace = True)

df_Parsivel['DateTime'] = df_Parsivel['Date'] + ' ' + df_Parsivel['Time']
df_Parsivel.drop(['Date','Time'], axis = 1, inplace = True)

df_Parsivel.index = df_Parsivel['DateTime'].apply(lambda x: datetime.strptime(x, '%d.%m.%Y %H:%M:%S'))
df_Parsivel.drop(['DateTime'], axis = 1, inplace = True)

df_Parsivel.sort_index(inplace=True)
df_Parsivel.index = df_Parsivel.index.floor('s')

df_Parsivel = df_Parsivel[~df_Parsivel.index.duplicated()]
reindxd_df = df_Parsivel.reindex(index=ind)
reindxd_df.index.name='Timestamp'

reindxd_df.sort_index(inplace=True)
reindxd_df.index = reindxd_df.index.floor('s')
````
