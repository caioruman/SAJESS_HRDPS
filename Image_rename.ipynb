Will rename images using the date and time from the image metadata 
#### Caution: A copy of the image is NOT created. The original files will be renamed.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
from PIL import Image
import piexif as pf
from PIL.ExifTags import TAGS
import time  
import glob
%matplotlib inline
```

```python
# To retrieve metadata from image file 

def get_exif(fn):
    ret = {}
    i = Image.open(fn)
    info = i._getexif()
    for tag, value in info.items():
        decoded = TAGS.get(tag, tag)
        ret[decoded] = value
    return ret

# Path the photos are located at
os.chdir('C:\\Users\\amand\\Documents\\PYTHON\\SAJESS\\microphoto\\3')

# File type of photos to be renamed 
files = glob.glob('*.JPG')

# For loop to grab all the image files 
for file in files:
    print(file)
    time = get_exif(file)['DateTimeOriginal']
    
# Formatting of the datetime 
    time = time.replace(':', '')
    time = time.replace(' ', '_')
    
# Add the following line to remove the seconds from the datetime
#    time = time[:-2]

# Add additional information after the datetime
    new_name = time+'_MT_MicroPhoto.jpg'
    if new_name == file:
        print(new_name, 'file is already renamed')

# Can use this ending if there is no chance of duplicated file names         
#    else:
#        os.rename(file, new_name)

# Use this ending to account for potential duplicates 
        continue
    while os.path.exists(new_name):
        new_name = time+'(1)'+'_MT_MicroPhoto.jpg'
    os.rename(file, new_name)
```
